```python
import nbformat as nbf

nb = nbf.v4.new_notebook()

nb.cells = [
    nbf.v4.new_markdown_cell(
        "# EDA — Bone Marrow Transplant\n**Coding Week 2026 — École Centrale Casablanca**"
    ),
    nbf.v4.new_code_cell(
        "import pandas as pd\n"
        "import numpy as np\n"
        "import matplotlib.pyplot as plt\n"
        "import seaborn as sns\n"
        "import warnings\n"
        "warnings.filterwarnings('ignore')\n\n"
        "df = pd.read_csv('../data/bone_marrow.csv')\n"
        "print(f'Dataset: {df.shape[0]} patients, {df.shape[1]} features')\n"
        "df.head()"
    ),
    nbf.v4.new_code_cell(
        "print('=== Dataset Info ===')\n"
        "df.info()\n"
        "print('\\nMissing values:')\n"
        "missing = df.isnull().sum()\n"
        "print(missing[missing > 0])"
    ),
    nbf.v4.new_code_cell(
        "plt.figure(figsize=(6,4))\n"
        "counts = df['survival_status'].value_counts()\n"
        "plt.bar(['Not Survived', 'Survived'], counts.values, color=['#e53935','#1e88e5'])\n"
        "plt.title('Survival Status Distribution')\n"
        "plt.ylabel('Count')\n"
        "for i, v in enumerate(counts.values):\n"
        "    plt.text(i, v+1, str(v), ha='center', fontweight='bold')\n"
        "plt.tight_layout()\n"
        "plt.show()\n"
        "print(f'Survived: {counts[1]} | Not Survived: {counts[0]}')"
    ),
    nbf.v4.new_code_cell(
        "numeric_df = df.select_dtypes(include=['number'])\n"
        "plt.figure(figsize=(14,10))\n"
        "sns.heatmap(numeric_df.corr(), cmap='RdBu_r', center=0, annot=False, linewidths=0.5)\n"
        "plt.title('Feature Correlation Heatmap')\n"
        "plt.tight_layout()\n"
        "plt.show()"
    ),
    nbf.v4.new_code_cell(
        "top_corr = numeric_df.corr()['survival_status'].abs().sort_values(ascending=False)[1:11]\n"
        "plt.figure(figsize=(8,5))\n"
        "top_corr.plot(kind='barh', color='#1e88e5')\n"
        "plt.title('Top 10 Features Correlated with Survival')\n"
        "plt.xlabel('Absolute Correlation')\n"
        "plt.tight_layout()\n"
        "plt.show()\n"
        "print(top_corr)"
    ),
    nbf.v4.new_code_cell(
        "numeric_cols = numeric_df.columns[:6].tolist()\n"
        "fig, axes = plt.subplots(2, 3, figsize=(14, 8))\n"
        "for i, col in enumerate(numeric_cols):\n"
        "    ax = axes[i//3][i%3]\n"
        "    for status, color, label in [(0,'#e53935','Not Survived'),(1,'#1e88e5','Survived')]:\n"
        "        subset = df[df['survival_status']==status][col].dropna()\n"
        "        ax.hist(subset, alpha=0.6, color=color, label=label, bins=20)\n"
        "    ax.set_title(col)\n"
        "    ax.legend(fontsize=8)\n"
        "plt.suptitle('Feature Distributions by Survival Status')\n"
        "plt.tight_layout()\n"
        "plt.show()"
    ),
    nbf.v4.new_markdown_cell(
        "## Summary\n"
        "- Dataset: 187 pediatric patients, 37 features\n"
        "- Target: `survival_status` (binary)\n"
        "- Key features: survival_time, CD34kgx10d6, ANCrecovery\n"
        "- Models: XGBoost (best), LightGBM, Random Forest, SVM"
    ),
]

with open('notebooks/eda.ipynb', 'w', encoding='utf-8') as f:
    nbf.write(nb, f)

print('✅ Notebook created successfully!')
```
