# EDA Bone Marrow Transplant

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../data/bone_marrow.csv')
print(df.shape)
df.head()
```

```python
df['survival_status'].value_counts().plot(kind='bar', color=['red','blue'])
plt.title('Survival Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
sns.heatmap(df.select_dtypes(include=['number']).corr(), cmap='RdBu_r', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
